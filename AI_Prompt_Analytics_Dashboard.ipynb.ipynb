```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
print ("Libery added successfully")
```

```python
np.random.seed(42)
```

```python
ai_tools = ["ChatGPT", "Gemini", "Claude", "Copilot", "Perplexity"]

categories = [
    "Python",
    "Machine Learning",
    "Data Science",
    "SQL",
    "Web Development",
    "Deep Learning",
    "Computer Vision",
    "NLP"
]

dates = pd.date_range("2026-07-01", periods=100)

data = {
    "Prompt_ID": range(1, 101),
    "Date": np.random.choice(dates, 100),
    "AI_Tool": np.random.choice(ai_tools, 100),
    "Category": np.random.choice(categories, 100),
    "Prompt_Length": np.random.randint(30, 400, 100),
    "Response_Time": np.round(np.random.uniform(1.0, 5.0, 100), 2),
    "Rating": np.random.randint(1, 6, 100),
    "Success": np.random.choice(["Yes", "No"], 100, p=[0.9, 0.1]),
    "Time_Spent": np.random.randint(5, 60, 100)
}

df = pd.DataFrame(data)

df.to_csv("ai_prompts.csv", index=False)

print("Dataset Created Successfully ✅")
```
